#By ID and Class

```python
class_data=soup.find("p",class_="section-subheading")
print(class_data)
```

```python
id_data=soup.find("p",id_="section-subheading")
print(id_data)
```

```python
lines=soup.find_all("p")
lines
```

```python
for l in lines:
  print(l.text)
```

```python
s=soup.find("div",class_="content")
print(s)
```

#Links

```python
links=soup.find_all("a")
links
```

```python
for l in links:
  print(l.get("href"))
```

#Images

```python
img=soup.find_all("img")
```

```python
img
```

```python
for i in img:
  print(i.get("src"))
```
